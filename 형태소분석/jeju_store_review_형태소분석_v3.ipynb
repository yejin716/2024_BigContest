```python
def usage_ratio(usage_list):
    포장_배달_유무 = 1 if '포장·배달 이용' in usage_list else 0
    # '포장·배달 이용' 제외하고 남은 리스트
    filtered_list = [x for x in usage_list if x in ['예약 없이 이용', '예약 후 이용']]

    total_count = len(filtered_list)
    예약_유 = usage_list.count('예약 후 이용')
    예약_무 = usage_list.count('예약 없이 이용')

    #비중
    예약_유_비중 = 예약_유 / total_count if total_count > 0 else 0
    예약_무_비중 = 예약_무 / total_count if total_count > 0 else 0

    return pd.Series([예약_유_비중, 예약_무_비중, 포장_배달_유무])

df[['예약_유', '예약_무', '포장_배달_유무']] = df['이용방법'].apply(usage_ratio)
df.head(5)
```

```python
from collections import Counter

# 각 행의 tokens에 대해 단어 빈도수 계산하고 상위 10개 단어만 추출
def get_top_nouns(tokens):
    if not tokens:  # 빈 리스트인 경우
        return []
    count = Counter(tokens)  # 빈도수 계산
    return [word for word, _ in count.most_common(20)]  # 단어만 반환

# 각 행에 대해 상위 10개 단어를 계산하여 새로운 컬럼에 저장
df['top_tokens'] = df['pos_tokens'].apply(get_top_nouns)
```
