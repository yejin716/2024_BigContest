```python
# 문자열에서 기호를 제거하고 리스트로 변환하는 함수
import re
def clean_and_convert_to_list(s):
    # 정규 표현식을 사용하여 기호 제거
    clean_string = re.sub(r"[\'\[\]]", "", s)  # '와 [ ] 제거
    # 쉼표로 분할하고 공백 제거 후 리스트로 변환
    return [item.strip() for item in clean_string.split(',') if item.strip()]

# 문자열을 리스트로 변환
df['동행자'] = df['동행자'].apply(clean_and_convert_to_list)
```

```python
from collections import Counter

# 각 행의 tokens에 대해 단어 빈도수 계산하고 상위 10개 단어만 추출
def get_top_nouns(tokens):
    if not tokens:  # 빈 리스트인 경우
        return []
    count = Counter(tokens)  # 빈도수 계산
    return [word for word, _ in count.most_common(1)]  # 단어만 반환

# 각 행에 대해 상위 10개 단어를 계산하여 새로운 컬럼에 저장
df['동행자_top_nouns'] = df['동행자'].apply(get_top_nouns)
```

```python
# 각 행의 여러 리스트 컬럼을 하나의 리스트로 묶는 함수
def combine_lists(row):
    combined = []
    for value in row:
        if isinstance(value, list):  # 값이 리스트인지 확인
            combined.extend(value)  # 리스트를 결합
    return combined

# 원하는 컬럼을 선택하여 combine_lists 함수를 적용
df['combined_list'] = df[['이용방법_top_nouns', '대기시간_top_nouns', '방문목적_top_nouns',
       '동행자_top_nouns']].apply(combine_lists, axis=1)
```

```python
from collections import Counter

# 각 행의 tokens에 대해 단어 빈도수 계산하고 상위 10개 단어만 추출
def get_top_nouns(tokens):
    if not tokens:  # 빈 리스트인 경우
        return []
    count = Counter(tokens)  # 빈도수 계산
    return [word for word, _ in count.most_common(10)]  # 단어만 반환

# 각 행에 대해 상위 10개 단어를 계산하여 새로운 컬럼에 저장
df['tokens'] = df['tokens'].apply(get_top_nouns)
```
