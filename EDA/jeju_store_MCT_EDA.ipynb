## 비중 컬럼 %로 바꿈

```python
ratio_col = ['월_이용건수비중', '화_이용건수비중', '수_이용건수비중', '목_이용건수비중', '금_이용건수비중', '토_이용건수비중',
       '일_이용건수비중', '05-11시_이용건수비중', '12-13시_이용건수비중', '14-17시_이용건수비중',
       '18-22시_이용건수비중', '23-04시_이용건수비중', '현지인_이용건수비중', '남성_이용건수비중',
       '여성_이용건수비중', '10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중',
       '60대이상_회원비중']

df[ratio_col] = df[ratio_col].apply(lambda x: x * 100).round(2)
```

```python
df.head()
```

## 파생변수 best_age 생성
- 가장 비중이 높은 나이

```python
age_col =  ['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중', '60대이상_회원비중']
df['best_age'] = df[age_col].idxmax(axis=1)
df['best_age'] = df['best_age'].replace({
    '10-20대_회원비중': '10-20대',
    '30대_회원비중': '30대',
    '40대_회원비중': '40대',
    '50대_회원비중': '50대',
    '60대이상_회원비중': '60대이상'
})


print(df[['best_age']].head())
```

```python
df.head(5)
```

## 파생변수 best_gender
- 가장 비중이 높은 성별 컬럼

```python
age_col =  ['남성_이용건수비중',
       '여성_이용건수비중']
df['best_gender'] = df[age_col].idxmax(axis=1)
df['best_gender'] = df['best_gender'].replace({
    '남성_이용건수비중': '남성',
    '여성_이용건수비중': '여성'
})

print(df[['best_gender']].head())
```
