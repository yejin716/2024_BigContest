```python
import os
import numpy as np
import pandas as pd
import torch
import chromadb

from transformers import AutoTokenizer, AutoModel
from langchain.vectorstores import Chroma
from langchain.embeddings.huggingface import HuggingFaceEmbeddings


# 임베딩 생성
# embeddings = np.array([embed_text(summary) for summary in df['요약']])

# 메타데이터 생성
metadata = [{'name': str(row['가맹점명']), 'summary': str(row['요약'])} for _, row in df.iterrows()]

# ChromaDB 저장소 설정
embedding_function = HuggingFaceEmbeddings(model_name='jhgan/ko-sroberta-multitask')
```

```python
# ChromaDB 저장소 설정
persist_directory = r'D:\Yebang\study\2024빅콘테스트\생성형AI분야\ChromaDB'

# ChromaDB 인스턴스 생성
vectorstore = Chroma(persist_directory=persist_directory, embedding_function=embedding_function)

# 벡터스토어에 임베딩과 메타데이터 저장
vectorstore.add_texts(texts=df['요약'].tolist(), metadatas=metadata, embeddings=embeddings.tolist())
```

```python
# 임베딩 생성 및 ChromaDB에 추가
for index, row in df.iterrows():
    summary = row['요약']
    name = row['가맹점명']
    
    # 임베딩 생성
    embedding = embedding_function.embed_documents([summary])[0]  # 배열에서 첫 번째 요소를 가져옴

    # 메타데이터 설정
    metadata = {
        "name": str(name),  # 문자열로 변환
        "summary": str(summary)  # 문자열로 변환
    }

    # ChromaDB에 추가
    vectorstore.add_texts(texts=[summary], metadatas=[metadata], embeddings=[embedding])
```
